```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn  
```

```python
loss = pd.read_csv("LvsD.csv")
Load = loss.iloc[:, -3].values
DisA = loss.iloc[:, -2].values
DisP= loss.iloc[:, -1].values
plt.plot(Load, DisA, label = 'Actual Displacement' ,color = 'red')
plt.plot(Load, DisP, label = 'Predicted Displacement' ,color = 'Blue' , linestyle='dashed')
plt.xlabel('Load')
plt.ylabel('Displacement')
plt.legend(loc='upper left')
plt.savefig("LoadvsDis.png", format="png", dpi=300)
```

```python
loss2 = pd.read_csv("LvsC.csv")
Dis = loss2.iloc[:, -3].values
CapA = loss2.iloc[:, -2].values
CapP= loss2.iloc[:, -1].values
plt.plot(Dis, CapA, label = 'Actual Capacitance' ,color = 'red')
plt.plot(Dis, CapP, label = 'Predicted Capacitance' ,color = 'Blue' , linestyle='dashed')
plt.xlabel('Displacement')
plt.ylabel('Capacitance x 1E14')
plt.legend(loc='upper left')
plt.savefig("DisvsCap.png", format="png", dpi=300)
```
